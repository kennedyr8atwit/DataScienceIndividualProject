```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

years = [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]
growth_percentage = [0, 2, 4, 6, 8, 10, 12, 14, 18, 20, 22]

plt.figure(figsize=(10, 6))
plt.plot(years, growth_percentage, marker='o', color='b', label="Projected Growth (%)")
plt.title("Projected Job Growth for Data Science-related Roles (2020-2030)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Growth Percentage", fontsize=12)
plt.xticks(years)
plt.yticks(range(0, 25, 2))
plt.grid(True)

plt.legend()
plt.show()
#U.S. Bureau of Labor Statistics (BLS) projections
```

```python
df = pd.read_csv('individualProjectData(Sheet1).csv')

# Cleaning, replace 'N/A' with None (so pandas can handle them as NaN values)
df_cleaned = df.replace({'N/A': None})

df_cleaned.iloc[:, 1:] = df_cleaned.iloc[:, 1:].apply(lambda x: x.str.rstrip('%').astype(float))
df_cleaned.set_index('Skills/Sources', inplace=True)
plt.figure(figsize=(12, 8))
df_cleaned.T.plot(kind='bar', figsize=(12, 8), width=0.8)
plt.title('Skills/Languages Popularity by Source', fontsize=16)
plt.xlabel('Source', fontsize=12)
plt.ylabel('Popularity (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Skills/Languages', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
# statista, Medium, 365
```
